```python
grouped_counts_th = region_df['TH'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)
grouped_counts_ec = region_df['EC'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)
grouped_counts_hc = region_df['HC'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)
grouped_counts_ag = region_df['AG'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)
grouped_counts_mt = region_df['MT'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)
grouped_counts_pfc = region_df['PFC'].groupby('major.celltype')['cogdx'].value_counts().unstack(fill_value=0)

name = ['TH', 'EC', 'HC', 'MT', 'PFC','AG']
```

```python
name = ['TH', 'EC', 'HC', 'MT', 'PFC','AG']
```

```python
region_group_counts_list = [grouped_counts_th, grouped_counts_ec, grouped_counts_hc, grouped_counts_mt, grouped_counts_pfc, grouped_counts_ag]
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

for j,i in enumerate(region_group_counts_list):
    print(i.head())
 
    grouped_counts = i.reset_index()
    
    # Melt the DataFrame for seaborn
    melted_counts = grouped_counts.melt(id_vars='major.celltype', var_name='cogdx', value_name='count')
    
    # Create a bar plot
    plt.figure(figsize=(20, 6))
    sns.barplot(data=melted_counts, x='major.celltype', y='count', hue='cogdx')
    plt.title(f'Count of cell type  based on cogdx for Region: {name[j]} ')
    plt.xlabel('Cell Type')
    plt.ylabel('Count')
    plt.legend(title='Cogdx (clinical)')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
